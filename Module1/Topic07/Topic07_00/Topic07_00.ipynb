# Topic 7 - 循环语句 - `while` 循环

本章我们就进入到了另一种程序控制结构：循环结构的学习，本章我们先学习 `while` 循环，下一章我们会学习 `for` 循环。

- **Topic 7.1 - `while` 循环的基本用法**：学习 `while` 循环的基本语法，以及和 `break`、`continue`、`else` 语句的结合使用。
- **Topic 7.2 - `while` 循环的进阶用法**：学习 `while` 循环的嵌套使用。
- **Topic 7.3 - `while` 无限循环的程序设计**：学习如何通过无限循环来实现软件设计的基础功能，并开发一个简单的动物园程序。

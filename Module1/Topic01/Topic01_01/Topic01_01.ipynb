# Topic 1.1 - Python 语言简介

## 1. Python 的起源与发展历程

Python 语言的起源：

- Python 语言是1989年由 Guido van Rossum 发明的（当时 Guido 只有33岁），灵感来源于教学语言 ABC - 由 Guido 参加设计的一种教学语言
- 当时 ABC 这种语言非常优美和强大，是**专门为非程序员设计的**，但是 ABC 语言并没有成功，Guido 认为是非开放造成的
- 因此 Guido 决心设计一种新的编程语言，继承 ABC 优雅简单的特点，但是要保持开放

Python语言的主要发展历程：

- Python 语言的第一个完整版由 Guido 于1991年发布，并且开源，全世界工程师都可以参与到 Python 的开发与改进中，此后，Python 凭借简洁语法和强大功能，逐渐在学术界与工业界流行
- 在经历近10年的发展完善后，Python 2 版本于2000年发布，Python 2 就已经是当时应用最广泛的计算机语言
- Python 3 版本于2008年发布，并且沿用至今，截至2016年，Python 3 版本及重要的第三方库已基本完善
- 截至目前，Python 最新版已经是3.10+，Python 2 版本早已停止更新但仍可使用，Python 1 版本已经在现代操作系统中不能使用
- 如今，Python 已成为全球最受欢迎的编程语言之一，在数据分析、人工智能、Web 开发等领域应用广泛

## 2. Python 的语言特点

Python 能发展成为当今最流行的语言，得益于它的以下优点：

- **简洁易学**：语法接近自然语言，代码量少，开发效率高。
- **设计哲学**：强调优雅、明确、简单，主张“用一种方法来完成一件事”。
- **生态丰富**：Python 有着海量的标准库与第三方库，涵盖数据处理、网络通信、人工智能、科学计算等各类应用场景。
- **跨平台性**：支持多种操作系统，几乎可以在任何环境中运行。
- **可扩展性**：能与 C/C++ 等语言结合，兼顾灵活性与性能。
- **社区活跃**：使用广泛，拥有庞大的开发者社区。

# Topic 5 - Python 组合数据类型

除了数字和字符串这两种基本数据类型，Python 还提供了多种组合数据类型，用于存储和操作多个数据元素，本章我们就来学习这些组合数据类型：

- **Topic 5.1 - 组合数据类型之列表**：学习 Python 中的列表（List）数据类型，及其基本操作。
- **Topic 5.2 - 组合数据类型之元组**：学习 Python 中的元组（Tuple）数据类型，及其基本操作。
- **Topic 5.3 - 组合数据类型之集合**：学习 Python 中的集合（Set）数据类型，及其基本操作。
- **Topic 5.4 - 组合数据类型之字典**：学习 Python 中的字典（Dictionary）数据类型，及其基本操作。
- **Topic 5.5 - 组合数据类型之间的转换**：学习不同组合数据类型之间的相互转换方法。

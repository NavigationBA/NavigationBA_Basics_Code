# Topic 8 - 循环语句 - `for` 循环

本章我们来学习另一种常用的循环语句 - `for` 循环，在数据科学与分析领域，`for` 循环的使用频率远高于 `while` 循环。

- **Topic 8.1 - `for` 循环的基本用法**：学习 `for` 循环的基本语法结构，以及使用 `for` 循环遍历常见的组合数据类型。
- **Topic 8.2 - `for` 循环的进阶用法**：学习 `for` 循环的嵌套用法，以及 `for` 循环与 `break`、`continue`、`else` 语句的结合使用。
- **Topic 8.3 - `for` 循环的高级用法**：学习递推表达式，以及 `enumerate()` 函数和 `zip()` 函数在 `for` 循环中的使用方法。

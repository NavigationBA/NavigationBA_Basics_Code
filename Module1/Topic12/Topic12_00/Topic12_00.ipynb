# Topic 12 - 模块与包

本章我们学习如何使用引用 Python 中的的模块与包，以及如何自定义模块与包，并且了解一些常用的内置模块与第三方包：

- **Topic 12.1 - 模块与包的引用**：学习引用 Python 中的模块与包的常用方法，并学习模块与包的区别，以及模块与包的组织结构。
- **Topic 12.2 - 自定义模块与包**：学习如何自定义模块与包，提高代码的复用性。
- **Topic 12.3 - 常用的内置模块与包（补充）**：介绍 Python 中一些常用的内置模块，包括 time 与 datetime。
- **Topic 12.4 - 常用的第三方模块与包（补充）**：学习如何使用 pip 工具安装第三方包，并介绍一些常用的第三方包，包括 joblib 与 tqdm。

# Topic 14 - 金融计算器项目 - Python 基础结课项目（补充）

本章我们开发一个更正式一些的项目：金融计算器项目，作为 Python 基础课程的结课项目，项目开发中将综合运用我们在 Python 基础这门课中学到的知识点，并使用正式的软件开发流程和方法，帮助大家更好地理解和掌握 Python 编程。

- **Topic 14.1 - 项目背景介绍与主程序框架搭建**：介绍项目的背景需求，根据需求规划项目的目录结构和主程序框架，并先实现基本的辅助功能。
- **Topic 14.2 - 表达式计算功能**：实现表达式计算模块，支持用户输入数学表达式并计算结果。
- **Topic 14.3 - 现金流计算功能**：实现现金流计算模块，支持根据输入的现金流计算净现值（NPV）和内部收益率（IRR）。
- **Topic 14.4 - 金钱时间价值功能**：实现金钱时间价值模块，支持计算未来值（FV）、现值（PV）、年金（PMT）、利率（R）、与期数（N）。
- **Topic 14.5 - 完善主程序与后续工作**：将各计算模块集成到主程序中，并且完善 README 文档。
- **Topic 14.6 - 金融计算器程序升级 - 版本 v2**：对金融计算器程序进行升级，包括改进用户输入与强化 IRR 计算功能。
- **Topic 14.7 - 金融计算器程序升级 - 版本 v3**：继续对金融计算器程序进行升级，增加用户设置功能，支持用户自定义小数点位数和金额符号等选项。

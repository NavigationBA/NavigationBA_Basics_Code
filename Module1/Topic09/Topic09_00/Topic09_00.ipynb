# Topic 9 - 自定义函数

本章我们将学习如何在 Python 中定义和使用自定义函数，这是提高代码复用性和可读性的关键手段。

- **Topic 9.1 - 函数的基本用法**：学习自定义函数的基本语法，以及参数与返回值的使用。
- **Topic 9.2 - 函数的进阶用法**：学习函数参数默认值、函数的作用域、以及函数注释。
- **Topic 9.3 - 函数的高级用法（补充）**：学习函数嵌套使用及其多层级的作用域、参数的可变与不可变类型，以及多值参数。
- **Topic 9.4 - 函数拓展（补充）**：了解如何定义与使用 `lambda` 函数，以及函数作为 Python 中对象的使用情景。

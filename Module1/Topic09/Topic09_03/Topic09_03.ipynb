### (1) åœ¨å‡½æ•°å†…éƒ¨å®šä¹‰å‡½æ•°

åœ¨ Python ä¸­ï¼Œæˆ‘ä»¬å¯ä»¥åœ¨å‡½æ•°å†…éƒ¨å®šä¹‰å¦ä¸€ä¸ªå‡½æ•°ï¼Œæˆ‘ä»¬æ¥çœ‹ä¸€ä¸ªç®€å•çš„ä¾‹å­ï¼š

````python
def calculate(a, b):

    def add(x, y):
        sum_result = x + y
        return sum_result

    def subtract(x, y):
        diff_result = x - y
        return diff_result

    sum_result = add(a, b)
    diff_result = subtract(a, b)
    
    return sum_result, diff_result
````

````python
a = 10
b = 5
result = calculate(a, b)
print(f"{a}ä¸{b}çš„å’Œä¸å·®åˆ†åˆ«æ˜¯:", result)
````

- é€šå¸¸æˆ‘ä»¬ç»™è¿™ä¸ªå‚æ•°èµ·åä¸º `*args`ï¼Œè¿™ä¸ªæ˜¯ä¸€ä¸ªçº¦å®šä¿—æˆçš„åå­—

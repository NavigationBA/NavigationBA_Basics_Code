# Topic 2 - Python 程序基本要素

本章我们将正式进入 Python 编程的学习，在这一章中，你将接触 Python 程序的几个最基本要素，这些内容将为后续的学习打下基础：

- **Topic 2.1 - 一些简单的Python程序**：我们将通关几个简单的，例如 `print("Hello World")` 这样的程序，来体验 Python 程序的运行原理
- **Topic 2.2 - 变量赋值**：学会用变量保存和管理数据的基本方法
- **Topic 2.3 - 注释**：用文字说明代码，让程序更容易理解和维护
- **Topic 2.4 - `input()` 函数**：获取用户输入，使程序具有交互性
- **Topic 2.5 - `print()` 函数拓展功能**：了解 `print()` 函数的更多用法，控制输出格式，让结果更清晰 
- **Topic 2.6 - Python 代码规范**（补充）：了解 Python 官方推荐的代码编写规范，养成良好的编程习惯

# Topic 2.4 - 用户输入 - `input()` 函数

在很多程序中，除了输出信息给用户，还需要接收用户的输入，让程序能够与用户进行交互。Python 提供了 `input()` 函数来实现这一功能。

由于 `input()` 函数在 `.ipynb` 文件中实在有些鸡肋，我们这一章主要在 `.py` 文件中给大家演示。

## 1. `input()` 函数的基本使用

在 Python 中，获取用户输入要使用内置函数 **`input()`**，基本语法是：  

```python
变量名 = input("提示信息：")
```

- 首先，`input()` 函数需要使用一个变量来接收
- 其次，`input()` 函数中可以输入一段文字作为提示信息
- 当 Python 解释器看到有 `input()` 函数的时候，就会在控制台等待用户输入，用户如果没有输入，程序会一直等待
- 用户可以输入一段文字在控制台，输入结束后，按回车结束，程序就会继续运行

我们尝试运行以下代码，大家可以尝试输入任意文字，例如我输入了“张三”：

```python
name = input("请输入你的名字：")
print(name)
```

```text
请输入名字：张三
张三
```

## 2. `input()` 返回值的类型

无论用户输入的是数字还是文字，`input()` 的返回值都是**字符串**类型：

- 比方说以下这个例子，我们在以下程序中输入“18”：

```python
age = input("请输入你的年龄：")
print(age)
print(type(age))
```

```text
请输入你的年龄：18
18
<class 'str'>
```

- 可以看到，`age` 变量虽然看起来是数字，但实际是字符串。

因此，如果想要将用户输入的数字真正变成 Python 中的数字类型，我们就需要**转换函数**：

| 函数         | 说明              |
| ----------  | ------------      | 
| `int(x)`    | 将 `x` 转换为整数   |
| `float(x)`  | 将 `x` 转换为浮点数 |

- 这次，我们将上面例子中的输入套入一个 `int()` 函数中，将它转换为 Python 中的整数类型：

```python
age_new = int(input("请输入你的年龄："))
print(age_new)
print(type(age_new))
```

```text
请输入你的年龄：18
18
<class 'int'>
```

- 此外，我们再尝试以下 `float()` 函数：

```python
height = float(input("请输入你的身高："))
print(height)
print(type(height))
```

```text
请输入你的身高：178.5
178.5
<class 'float'>
```

这里我们注意额外两点：

- `int()` 函数与 `float()` 函数中，只能输入“看上去像是数字的文字”，如果输入非数字，就会报错，例如 `int("abc")` 就会报错

- 如果用户在控制台什么也不输入就按下回车，那么 `input()` 函数接收到的是一个**空字符串**

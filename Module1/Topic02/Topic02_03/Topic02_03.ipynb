# Topic 2.3 - 注释

在编程中，**注释（comments）**是写在代码里的说明文字，用来帮助程序员理解代码逻辑

- 注释不会被 Python 解释器执行，只是为了让人更好地阅读代码
- 在 Python 中有两种注释方式，**单行注释** 和 **多行注释**



## 1. 单行注释

单行注释以井号 `#` 开头，井号右边的内容就是注释，可以写在一行的开头，也可以跟在代码的后面

````python
# 以下代码展示了 print 函数的使用
print("Hello World!")  # 输出 Hello World
````

Python 解释器在处理注释时的原理是：

- Python 解释器在看到井号以后，就知道井号后面是注释了，就自动跳过井号及其后面的内容了
- 可以看到，Python 只运行了 `print()` 函数，自动忽视了那两行的注释

以井号开头的注释，还有一个常用的用途，就是用来**临时屏蔽掉某行代码**，让这行代码不被执行

- 这种用法在调试程序时非常有用，可以快速地禁用某些代码行，而不需要删除它们
- 其实有一个快捷键：选中要注释的代码后，Mac 上是 `Command + /`，Windows 上是 `Ctrl + /`，可以快速地给选中的代码行添加或删除注释符号 `#`

````python
# print("欢迎学习Python编程！")  # 该行代码被注释掉了，不会执行
print("让我们一起探索编程的乐趣吧！")
print("祝你学习愉快！")
````

## 2. 多行注释
多行注释适用于文字比较长，使用三个连续的双引号，或三个连续的单引号：

```python
"""
注释内容
"""

'''
注释内容
'''
```

````python
"""
以下的代码会：
1. 输出Hello World!
2. 输出Hello Python!
"""

print("Hello World!")
print("Hello Python!")
````

使用三个引号的多行注释，与使用井号定义的单行注释，在本质上是不同的：

- 多行注释的本质是多行字符串，字符串是 Python 中的一种数据类型，就是 Python 中的文字
- Python 解释器看到多行注释后，会发现这里有个多行字符串，并从头读到尾
- 但由于这个多行字符串并没有做任何处理或输出，所以 Python 解释器会将它单纯地摆在这里，不做任何操作
- 因此，多行字符串可以当做多行注释来使用
